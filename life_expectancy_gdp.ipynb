```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('all_data.csv')
df.replace(to_replace='United States of America', value='USA', regex= True, inplace=True)
print(df.head())
print(df.Country.value_counts())
```

```python
#ideas for comparison: life expectancy compared to GDP per country? for now lets just go with that
```

```python
sns.scatterplot(data=df, x=df.Country, y=df['Life expectancy at birth (years)'], hue = df.Year)
plt.title('Life Expectancy Over Time by Country')
plt.savefig('lifeexpectancycountryscatter.png')
plt.show()
plt.clf()
```

```python
sns.displot(data=df, x=df.Country, y=df.GDP, hue=df.Year)
plt.title('GDP Over Time by Country')
plt.savefig('countryGDPdisplot.png')
plt.show()
plt.clf()
```

```python
#correlations seems to be pretty minimal, meaning other factors are more significant (presence of social programs?)
#life expectancy can be clearly seen to be increasing over time in all countries in the dataframe.
#zimbabwe and chile have similar GDPs but massively different life expectations, suggesting low GDP is not a deciding factor
```
